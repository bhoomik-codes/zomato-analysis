# Phase 4: Model Preparation & Encoding

In this notebook, we transform our cleaned data into a format that Machine Learning models can understand.

```python
import pandas as pd
import sys
import os
import numpy as np

# Add src to path
sys.path.append(os.path.abspath('../'))
from src.data_cleaning import handle_rate, bin_price, encode_categorical
```

## Step 1: Load and Clean Initial Data

Perform the necessary cleaning steps from Phase 1.

```python
df = pd.read_csv('../data/Zomato-data-.csv')
df['rate'] = df['rate'].apply(handle_rate)
df['price_range'] = df['approx_cost(for two people)'].apply(bin_price)
print(f"Data loaded with {df.shape[0]} rows and {df.shape[1]} columns.")
```

## Step 2: Apply Categorical Encoding

We use Label Encoding for binary features (Yes/No) and One-Hot Encoding for the restaurant types.

```python
df_ml = encode_categorical(df)
print("Top 5 rows of encoded dataset:")
df_ml.head()
```

> _**Conclusion**: Categorical variables like 'online_order' and 'book_table' have been converted to 0/1, and restaurant types have been expanded into binary indicator columns (One-Hot Encoding)._

## Step 3: Select Features for Modeling

Remove non-numeric columns that can't be used directly by regression models.

```python
# Filter for numeric features and the target 'rate'
final_features = df_ml.select_dtypes(include=[np.number])
print(f"Final features selected: {final_features.columns.tolist()}")

# Save for modeling phase
final_features.to_csv('../data/Zomato_Processed_ML.csv', index=False)
print("Processed dataset saved to data/Zomato_Processed_ML.csv")
```
